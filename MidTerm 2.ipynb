```python
import matplotlib.pyplot as plt
import numpy as np

class polynomial(object):
    def __init__(self, c, v):
        self.coeff = c
        self.v = v
        #df
        self.degree = len(c) - 1
    
    def __repr__(self):
        D = len(self.coeff)
        for i in range(D):
            if i<D-1:
                print(self.coeff[i], end='')
                print('{}^{} '.format(self.v, D-i-1), end='')
            else:
                print(self.coeff[i])
        
        return str(self.coeff)
    
    #Add Polynoms
    
    def __add__(self,b):
        """Computes a+b and returns the result"""
        d = max(self.degree, b.degree) + 1 
        
        #Equal to higher degree by adding "0"s
        
        self_mod =  [0]*(d-self.degree-1)+self.coeff 
        b_mod = [0]*(d-b.degree-1)+b.coeff

        coeff = []
        for i in range(d):
            coeff.append(self_mod[i] + b_mod[i])
        
        return polynomial(coeff, self.v)
    
    #Substract Polynoms
    
    def __sub__(self,b):
        """Computes a-b and returns the result"""
        d = max(self.degree, b.degree) + 1 #Max length of polynomials
        
        self_mod =  [0]*(d-self.degree-1)+self.coeff
        b_mod = [0]*(d-b.degree-1)+b.coeff
        
        coeff = []
        for i in range(d):
            coeff.append(self_mod[i] - b_mod[i])
        
        return polynomial(coeff, self.v)
    
    #Multiply Polynoms
    
    def __mul__(self,b):
        """Computes a*b and returns the result"""
        
        
        n = self.degree + b.degree     
        mul_coeff = [0]*(n + 1)
        
        for i in range(0, self.degree + 1):
            for j in range(0, b.degree + 1):
                mul_coeff[i+j] += self.coeff[i] * b.coeff[j]
        
        return polynomial(mul_coeff, self.v)
    
    #Polynom Graph
    
    def plot(self, number_of_points = 1000):
        plot_range = round(number_of_points / 2) + 1
       
        for i in range(plot_range):
            plt.plot(i, np.polyval(self.coeff, i), marker='.')
            plt.plot(-i, np.polyval(self.coeff, -i), marker='.')
        
        
        plt.xlabel('x', fontsize=16)
        plt.ylabel('y', fontsize=16)
        plt.show()


p = polynomial([2,3,1], 'z')
print(p)


q = polynomial([4,4], 'z')
print(q)

print('--------------------')
print('Result of p + q:')
r1 = p + q
print(r1)

print('--------------------')
print('Result of p - q:')
r2 = p - q
print(r2)

print('--------------------')
print('Result of p * q:')
r3 = p * q
print(r3)

print('--------------------')

p.plot()

q.plot()
```
